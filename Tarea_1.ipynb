## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director (
    id_director SERIAL PRIMARY KEY,
    nombre VARCHAR(45),
    edad INT
);
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor (
    id_actor SERIAL PRIMARY KEY,
    nombre VARCHAR(45),
    edad INT
);
```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (
    id_pelicula SERIAL PRIMARY KEY,
    id_director INT,
    nombre VARCHAR(45),
    genero VARCHAR(45),
    duracion INT,
    fecha_publicacion DATE,
    UNIQUE (id_pelicula) 
);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (
    id_actor INT,
    id_pelicula INT,
    PRIMARY KEY (id_actor, id_pelicula),
    FOREIGN KEY (id_actor) REFERENCES actor(id_actor),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (
    id_premio SERIAL,
    id_director INT,
    anio INT,
    PRIMARY KEY (id_premio, id_director),
    FOREIGN KEY (id_director) REFERENCES director(id_director)
);
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (
    id_premio SERIAL,
    id_actor INT,
    anio INT,
    PRIMARY KEY (id_premio, id_actor),
    FOREIGN KEY (id_actor) REFERENCES actor(id_actor)
);
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula (
    id_premio SERIAL PRIMARY KEY,
    id_pelicula INT,
    anio INT,
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)
);
```
